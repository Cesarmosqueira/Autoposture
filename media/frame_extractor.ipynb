```python
import cv2
import matplotlib.pyplot as plt
from PIL import Image
import random
```

If there are any errors, run a couple times 

```python
def get_two_random_numbers(n):
    number1 = random.randint(0, n)
    number2 = random.randint(0, n)
    while number2 == number1:
        number2 = random.randint(0, n)
    return number1, number2
```

```python
cap = cv2.VideoCapture('frani_pose.mp4')
```

```python
if (cap.isOpened()== False): 
  print("Error opening video stream or file")


i, l = 0, int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

frame1, frame2 = get_two_random_numbers(l)
images = []

success = cap.grab()
fno = 0
while success:
    if fno % 1 == 0:
        _, img = cap.retrieve()
        if i == frame1 or i == frame2:
            images += [cv2.cvtColor(img, cv2.COLOR_BGR2RGB)]
    # read next frame
    success = cap.grab()
    i += 1
```

```python
# Save the image as JPEG
Image.fromarray(images[0]).save('frani_pose1.jpg')
Image.fromarray(images[1]).save('frani_pose2.jpg')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

# Plot the first image
axes[0].imshow(images[0])
axes[0].axis('off')
axes[0].set_title("frani_pose1.jpg")

# Plot the second image
axes[1].imshow(images[1])
axes[1].axis('off')
axes[1].set_title("frani_pose2.jpg")

plt.show()
```
